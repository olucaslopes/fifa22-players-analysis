```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./preprocessing/data/filled_data_2022-06-21.tsv', sep='\t')
df.head()
```

```python
bool_cols = [col for col in df if 
               df[col].dropna().value_counts().index.isin([0,1]).all()]
bool_cols
```

```python
features = df.columns[:171][~df.columns[:171].str.match('(.*_id.*|.*id_.*|.*url.*)')].to_list()
features.remove('#FK Specialist')
new_features = df[features]._get_numeric_data()
new_features
```

```python
X, y = df._get_numeric_data().drop(columns=['#FK Specialist']), df['#FK Specialist']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import RFECV
# define the method
selector = RFECV(scoring= 'recall',estimator=LogisticRegression(), n_jobs=-1)
# fit the model
selector = selector.fit(new_features, y)
# Report feature ranking
order = pd.DataFrame({'coluna': X.columns}, index=selector.ranking_).sort_index().rename_axis('prioridade', axis=1)
order
```
